##Data Preprocessing
Preprocess the data to split sender, receiver, subject, message body, cc, bcc

```python
#Creating the function to split the message into sender receiver subject message body, cc, bcc

def email_parsing(raw_message):
    lines = raw_message.split('\n')
    email = {}
    message = ''
    keys_to_extract = ['subject','from', 'to', 'x-cc', 'x-bcc','x-folder', 'x-origin']
    for line in lines:
        if ':' not in line: #main body does not contain :
            message += line.strip()
            email['body'] = message
        else:
            pairs = line.split(':', 1) #Split first occurrance of ':'
            key = pairs[0].lower()
            val = pairs[1].strip()
            if key in keys_to_extract:
                email[key] = val
    return email

def emails_parsing(messages):
    emails = [email_parsing(message) for message in messages]
    return {
        'subject': result_append(emails, 'subject'),
        'to': result_append(emails, 'to'),
        'from': result_append(emails, 'from'),
        'cc': result_append(emails, 'x-cc'),
        'bcc': result_append(emails, 'x-bcc'),
        'body': result_append(emails, 'body'),
        'folder': result_append(emails, 'x-folder'),
        'origin': result_append(emails, 'x-origin')
    }


def result_append(emails, key):
    results = []
    for email in emails:
        if key not in email:
            results.append('')
        else:
            results.append(email[key])
    return results
```

```python
category_keywords = {
    'Spam': ['fun emails', 'junk e-mail', 'newsletter', 'motley fool', 'stored messages','junk', 'junk file', 'motley fool e-mails', 'eBay', 'advertising'],
    'HR': ['salary', 'promotion', 'payroll', 'benefits', 'hiring', 'schedule', 'hr'],
    'Resume': ['resume', 'cv', 'job application', 'candidate', 'recruiting', 'resumes'],
    'Finance': ['credit', 'bankruptcy', 'collateral', 'charges', 'billing', 'refund', 'bankrupt', 'citibank'],
    'Accounting': ['accounting', 'financial reporting', 'finance study group'],
    'Legal': ['lawsuit', 'litigation', 'compliance', 'regulation', 'settlements', 'risk management'],
    'Marketing': ['campaign', 'branding', 'promotion'],
    'Products': ['pipeline', 'gas', 'fuel', 'energy', 'natural gas', 'risk analytics']
}

def categorize_email(folder):
    folder = folder.lower()  # Convert to lowercase for case-insensitive matching
    for category, keywords in category_keywords.items():
        if any(keyword in folder for keyword in keywords):
            return category
    return 'Uncategorized'  # Default category if no match is found

filtered_data['category'] = filtered_data['folder'].apply(categorize_email)
```

```python
import plotly.express as px

def create_word_frequency_chart(hierarchy_unique, hierarchy_count, order = None):
  #prepare data for Plotly
  plot_hierarchy = pd.DataFrame({
      'Category': hierarchy_unique,
      'Frequency': hierarchy_count
  })

  #sort based on the value of order
  if order == 'descending':
    plot_hierarchy = plot_hierarchy.sort_values(by='Frequency', ascending=False)
  elif order == 'ascending':
    plot_hierarchy = plot_hierarchy.sort_values(by='Frequency', ascending=True)

  #create interactive bar chart
  fig = px.bar(plot_hierarchy, x='Category', y='Frequency', title='Interactive Word Frequency Chart')

  #display the chart
  fig.show()
```

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained model
model = SentenceTransformer("all-MiniLM-L6-v2")

#turn numpy array to python array

hierarchy_1_unique_list = hierarchy_1_unique.tolist()

# Get embeddings for your terms
embeddings = model.encode(hierarchy_1_unique_list)
```

```python
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Perform hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=15, linkage='ward')
clusters_hierarchical = hierarchical.fit_predict(embeddings)

# Print clusters
for i in range(15):
  cluster_i = []
  for term, cluster in zip(hierarchy_1_unique_list, clusters_hierarchical):
    if cluster == i:
      cluster_i.append(term)
  print(f"Cluster {i}: {cluster_i}")


# for term, cluster in zip(hierarchy_1_unique_list, clusters_hierarchical):
#     print(f"Term: {term}, Cluster: {clusters_hierarchical}")


# Reduce dimensionality for visualization
tsne = TSNE(n_components=2, random_state=42)
reduced_embeddings = tsne.fit_transform(embeddings)

# Create a DataFrame for visualization
df = pd.DataFrame({
    'x': reduced_embeddings[:, 0],
    'y': reduced_embeddings[:, 1],
    'term': hierarchy_1_unique_list,
    'cluster': clusters_hierarchical
})

# Interactive plot with Plotly
fig = px.scatter(
    df,
    x='x',
    y='y',
    color='cluster',
    hover_data=['term'],
    title='Interactive Clustering Visualization',
    labels={'color': 'Cluster'}
)

fig.show()
```

```python
from sklearn.cluster import SpectralClustering
from sklearn.manifold import TSNE


# Perform spectral clustering
spectral = SpectralClustering(n_clusters=15, affinity='nearest_neighbors', random_state=42)
clusters_spectral = spectral.fit_predict(embeddings)

# Print clusters
for i in range(15):
  cluster_i = []
  for term, cluster in zip(hierarchy_1_unique_list, clusters_spectral):
    if cluster == i:
      cluster_i.append(term)
  print(f"Cluster {i}: {cluster_i}")



# Reduce dimensionality for visualization
tsne = TSNE(n_components=2, random_state=42)
reduced_embeddings = tsne.fit_transform(embeddings)

# Create a DataFrame for visualization
df = pd.DataFrame({
    'x': reduced_embeddings[:, 0],
    'y': reduced_embeddings[:, 1],
    'term': hierarchy_1_unique_list,
    'cluster': clusters_spectral
})

# Interactive plot with Plotly
fig = px.scatter(
    df,
    x='x',
    y='y',
    color='cluster',
    hover_data=['term'],
    title='Interactive Clustering Visualization',
    labels={'color': 'Cluster'}
)

fig.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score


def evaluate_clustering(embeddings, clusters):
  # Calculate Silhouette Score
  silhouette_avg = silhouette_score(embeddings, clusters)
  print(f"Silhouette Score: {silhouette_avg}")
  # Calculate Calinski-Harabasz Index
  calinski_harabasz = calinski_harabasz_score(embeddings, clusters)
  print(f"Calinski-Harabasz Index: {calinski_harabasz}")
  # Calculate Davies-Bouldin Index
  davies_bouldin = davies_bouldin_score(embeddings, clusters)
  print(f"Davies-Bouldin Index: {davies_bouldin}")

print("hierarchical:")
evaluate_clustering(embeddings, clusters_hierarchical)
print("spectral:")
evaluate_clustering(embeddings, clusters_spectral)
```

Spectral Clustering for Hierarchy 2 & 3

```python
# Define remaining hierarchies
remaining_hierarchies = {
    "Hierarchy 2": hierarchy_2_unique.tolist(),
    "Hierarchy 3": hierarchy_3_unique.tolist()
}

# Dictionary to store clustering results
cluster_results = {}

# Process remaining hierarchies
for name, term_list in remaining_hierarchies.items():
    print(f"\n🔹 Processing {name}...")

    # Get sentence embeddings
    embeddings = model.encode(term_list)

    # Perform Spectral Clustering
    spectral = SpectralClustering(n_clusters=15, affinity='nearest_neighbors', random_state=42)
    clusters = spectral.fit_predict(embeddings)

    # Print Clusters
    print(f"\nClusters for {name}:")
    for i in range(15):
        cluster_i = [term for term, cluster in zip(term_list, clusters) if cluster == i]
        print(f"Cluster {i}: {cluster_i}")

    # Reduce dimensionality using t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    reduced_embeddings = tsne.fit_transform(embeddings)

    # Create DataFrame for visualization
    df = pd.DataFrame({
        'x': reduced_embeddings[:, 0],
        'y': reduced_embeddings[:, 1],
        'term': term_list,
        'cluster': clusters
    })

    # Store clustering results
    cluster_results[name] = df

    # Plot clustering result
    fig = px.scatter(
        df,
        x='x',
        y='y',
        color='cluster',
        hover_data=['term'],
        title=f'{name} Clustering Visualization',
        labels={'color': 'Cluster'}
    )

    fig.show()
```
